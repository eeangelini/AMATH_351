```python
# import necessary python packages
import numpy as np
import matplotlib.pyplot as plt

# make it so that plots will appear in-line in your notebook (i.e., next to the code)
%matplotlib inline 
```

Suppose we want to plot the <i> level curves </i> of a function $f(x,y)$. That is to say, we want to plot the implicit curve $f(x,y)=C$ for various values of $C$. This notebook will show you how to do so in Python using the example
$$
f(x,y) = x^2 - \frac12x^4 - y^2
$$

#### Step 1: Define your function $f$
In order to plot the level curves of our function $f(x,y)$, we must first define a Python function that takes values of `x` and `y` as input, and outputs the value of `x**2 - 0.5*x**4 -y**2`. In Python, the syntax for exponents is `**`, e.g., we write $x^2$ as `x**2`. We do so as follows.

```python
def f(x,y):
    return x**2 -0.5*x**4 - y**2
```

#### Step 2: Define the numerical range of $x$ and $y$
Once we have defined our function `f(x,y)`, we must specify the range of `x` and `y` over which we are going to plot the level curves of $f$. In particular, we must specify them as an arrays of points
$$
(x_0, x_1, x_2,\dots, x_n), \quad (y_0, y_1, y_2, \dots, y_m),
$$
so that we can plot $f(x,y) = C$ at each pair $(x_i, y_j)$.

In Python, we do so using the `numpy.linspace` function, abbreviated here as `np.linspace` (as we imported the module `numpy` under the alias `np`). This function takes as input `start, stop, num`, and returns a 1D array of `num` evenly spaced points going from `start` to `stop`. The number `num` of points in our array will determine how refined our numerical plot will be. In this example, we want to plot the level sets for $-2\leq x \leq 2$ and $-2 \leq y \leq 2$, and will do so over `num`=100 points on each interval.

```python
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
```

#### Step 3: Create a mesh grid over $x$ and $y$
In order to numerically plot the level curves for values of $x$ and $y$ specified in step 2, we must make a rectangular <i> grid </i> over those values. That is, given two 1D arrays `x` and `y` that contain our range of $x$ and $y$ values, respectively, we want to create a 2D grid (`X`,`Y`) such that the point $(x_i, y_j) = $ (`x[i]`,`y[j]`) corresponds to indexing our 2D arrays `X` and `Y` at `[i,j]`: (`X[i,j]`,`Y[i,j]`).  

In Python, we do so using the `numpy.meshgrid` function, abbreviated here as `np.meshgrid`. This function takes 1D arrays `x` (with length `N`) and `y` (with length `M`) as input, and returns the appropriate 2D arrays with `N` rows and `M` columns.

```python
X,Y = np.meshgrid(x,y)
```

#### Step 4: Plot your level curves

```python
plt.figure(figsize=(7,6)) # set the dimensions of your figures

myLevels = [-3.25,-2,-1,-0.25,0,0.25,0.4] # values of C for f(x,y) = C, must be an increasing list

plt.contour(X,Y,f(X,Y),myLevels,linewidths=3) # plot level curves

plt.xlabel('x',fontsize=14) # label x axis
plt.ylabel('y',fontsize=14) # label y axis
plt.xticks(fontsize=12); # set fontsize for markers on x axis
plt.yticks(fontsize=12); # set fontsize for markers on y axis

plt.savefig('level_curves_f.png',dpi=400) # save your figure as a png with resolution 400 dpi
```
