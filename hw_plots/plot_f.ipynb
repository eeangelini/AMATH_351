```python
# import necessary python packages
import numpy as np
import matplotlib.pyplot as plt

# make it so that plots will appear in-line in your notebook (i.e., next to the code)
%matplotlib inline 
```

Suppose we want to plot the graph of a function $f(x)$. That is to say, we want to plot the points $(x, f(x))$. This notebook will show you how to do so in Python using the example
$$
f(x) = \cos(\pi x + \pi/2)
$$

#### Step 1: Define your function $f$
In order to plot the graph of our function $f(x)$, we must first define a Python function that takes values of `x` and outputs the value of `cos(pi*x + pi/2)`. The function `cos` and the constant `pi` are already defined for us numerically in the package NumPy; we can call them using `np.cos()` and `np.pi`, respectively (recall that we imported the module `numpy` under the alias `np`).

```python
def f(x):
    return np.cos(np.pi*x + np.pi/2)
```

#### Step 2: Define the numerical range of $x$
Once we have defined our function `f(x)`, we must specify the range of `x` over which we are going to plot the graph of $f$. In particular, we must specify it as an arrays of points
$$
(x_0, x_1, x_2,\dots, x_n)
$$
so that we can plot $(x_i, f(x_i))$ at each value $x_i$.

In Python, we do so using the `linspace` function from NumPy, called here as `np.linspace`. This function takes as input `start, stop, num`, and returns a 1D array of `num` evenly spaced points going from `start` to `stop`. The number `num` of points in our array will determine how refined our numerical plot will be. In this example, we will to plot the graph for $-2 \pi - \pi/2 \leq x \leq 2\pi - \pi/2$, and will do so over `num`=250 points on this interval.

```python
x = np.linspace(-2*np.pi - np.pi/2,2*np.pi - np.pi/2,250)
```

#### Step 3: Plot your graph

```python
plt.figure(figsize=(7,6)) # set the dimensions of your figures

plt.plot(x,f(x),'k-',linewidth=3) # x, f(x), line color and style ('k-' = black, straight line), line width

plt.xlabel('x',fontsize=14) # label x axis
plt.ylabel('f(x)',fontsize=14) # label y axis
plt.xticks(fontsize=12); # set fontsize for markers on x axis
plt.yticks(fontsize=12); # set fontsize for markers on y axis

plt.title(r'Plot of $f(x) = \cos(2\pi + \pi/2)$',fontsize = 14)

plt.savefig('plot_f.png',dpi=400) # save your figure as a png with resolution 400 dpi
```
