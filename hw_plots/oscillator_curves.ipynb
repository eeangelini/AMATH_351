## Undamped driven harmonic oscillator & resonance

We want to plot our solution curves for the IVP from Problem 4 of HW4, which described the motion of an undamped oscillator with fundamental frequency $\omega_0 > 0$ driven by some external force at angular frequency $\omega > 0$:
$$
u'' + \omega_0^2u = \cos(\omega t), \ \ u(0) = 0, \ u'(0) = 0 .
$$
In the case that $\omega \ne \omega_0$, the solution was:
$$
u(t) = \frac{1}{\omega_0^2 - \omega^2} \Big(\cos(\omega t) - \cos(\omega_0 t) \Big) \, .
$$
In the case that $\omega = \omega_0$, the solution was
$$
u(t) = \frac{t}{2\omega_0}\sin(\omega_0 t) \, .
$$

```python
def u1(t,w,w0):
    return (np.cos(w*t) - np.cos(w0*t))/(w0**2 - w**2)

def u2(t,w0):
    return t*np.sin(w0*t)/(2*w0)
```

```python
def A1(t,w,w0):
    w1 = (w0-w)/2
    return 2*np.sin(w1*t)/(w0**2-w**2)

def A2(t,w0):
    return t/(2*w0)
```
